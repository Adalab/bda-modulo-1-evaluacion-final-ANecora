```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []
        self.ventas_totales = 0
        self.clientas = {}
#1
    def agregar_producto(self, nombre, precio, cantidad):

        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                producto["cantidad"] += cantidad
                print(f"El siguiente producto: {nombre}, ha actualizado su cantidad a {producto['cantidad']}")
                return
        else:
            producto_nuevo = {
                "nombre": nombre,
                "precio": precio,
                "cantidad": cantidad
            }
            self.inventario.append(producto_nuevo)
            print(f"Producto añadido: {producto_nuevo['nombre']}, Precio: ${producto_nuevo['precio']}, Cantidad: {producto_nuevo['cantidad']}")
#2
    def ver_inventario(self):

        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
#3
    def buscar_producto(self, nombre):

        for producto in self.inventario:
            if producto['nombre'].lower() == nombre.lower():
                print(f"Aquí tienes el producto buscado: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return
            
        print(f"Este producto: {nombre}, no está en el inventario.")
#4
    def actualizar_stock(self, nombre, cantidad):

        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                producto["cantidad"] += cantidad
                if producto["cantidad"] < 0:
                    producto["cantidad"] = 0
                print(f"Este es el stock actualizado: {nombre}, con su nueva cantidad {producto['cantidad']}")
                return
        print(f"Este producto: {nombre}, no está en el inventario.")
#5
    def eliminar_producto(self, nombre):
        
        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                self.inventario.remove(producto)
                print(f"El siguiente producto {nombre}, ha sido eliminado del inventario.")
                return
        print(f"Este producto: {nombre}, no está en el inventario.")
#6
    def calcular_valor_inventario(self):

        valor_inventario_total = 0

        for producto in self.inventario:
            valor_producto = producto["precio"] * producto["cantidad"]
            valor_inventario_total += valor_producto

        print(f"El valor total del inventario es de: ${valor_inventario_total}")

#BONUS
#1
    def agregar_clienta(self, nombre_clienta, email):

        if nombre_clienta in self.clientas:
            print(f"La clienta: {nombre_clienta} ya está registrada.")
        else:
            self.clientas[nombre_clienta] = {"nombre": nombre_clienta, "email": email}
            print(f"La clienta: {nombre_clienta} con el siguiente email: {email}, ha sido añadida correctamente.")
#2
    def ver_clientas(self):

        if not self.clientas:
            print("No hay clientas en la lista.")
        else:
            for clave, datos in self.clientas.items():
                print(f"Nombre: {datos['nombre']}, Email: {datos['email']}")
```

```python
tienda = TiendaOnline()

print("\n--Agregar productos al inventario--\n")
tienda.agregar_producto("Camisa", 20, 50)
tienda.agregar_producto("Pantalón Vaquero", 30, 30)
tienda.agregar_producto("Zapatos", 50, 40)
tienda.agregar_producto("Camisa", 20, 50)

print("\n--Buscar productos--\n")
resultado = tienda.buscar_producto("Camisa")
resultado = tienda.buscar_producto("Sombrero")

print("\n--Actualizar stock--\n")
tienda.actualizar_stock("Camisa", -13)
tienda.actualizar_stock("Zapatos", 30)
tienda.actualizar_stock("Sudadera", 10)

print("\n--Agregar y actualizar producto--\n")
tienda.agregar_producto("Sudadera", 30, 40)
tienda.actualizar_stock("Sudadera", 40)

print("\n--Eliminar producto--\n")
tienda.eliminar_producto("Zapatos")
resultado = tienda.buscar_producto("Zapatos")

print("\n--Agregar más productos al inventario--\n")
tienda.agregar_producto("Zapatos", 50, 50)
tienda.agregar_producto("Bufanda", 10, 60)
tienda.agregar_producto("Guantes", 18, 40)
tienda.agregar_producto("Camiseta Tirantes", 20, 38)
tienda.agregar_producto("Sombrero", 35, 25)
tienda.agregar_producto("Pantalón Corto", 24, 16)
tienda.agregar_producto("Vestido", 30, 43)

print("\n--Ver inventario actual--\n")
tienda.ver_inventario()

print("\n--Calcular valor total del inventario--\n")
tienda.calcular_valor_inventario()
```
